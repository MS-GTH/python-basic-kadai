```python

class Human:
    def __init__(self,name,age):
      self.name = name
      self.age = age

    def check_adult(self):
      if self.age >= 20:
        print("大人")
      else:
        print("大人でない")

man1 = Human("Ishikawa",21)
man2 = Human("Togusa",11)
woman1 = Human("Kusanagi",20)
woman2 = Human("Esaki",2)
staff = Human("Tachikoma",100)

Instance_lists = [man1,man2,woman1,woman2,staff]

count = len(Instance_lists)

for i in range(0, count):
  Instance_lists[i].check_adult()


```
